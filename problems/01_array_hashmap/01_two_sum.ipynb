# LeetCode 1: Two Sum

## Pattern
HashMap / Dictionary

## Difficulty
Easy

## Problem

Given an array of integers `nums` and an integer `target`, return the `indices` of the two numbers such that they add up to `target`.

Example:

```python
nums = [2, 7, 11, 15]
target = 9
```

## My First Thought: Brute Force

My first idea was to use a loop. For each number, I can compare it with every other number and check whether the sum equals the target.

## Complexity of Brute Force

If there are `n` numbers, the brute-force method compares many pairs.

The number of comparisons is approximately: n(n - 1) / 2
The time complexity is: O(n²)

## Complexity of HashMap
Time: O(n), as Hashmap complexity is O(1), and we loop n---n × O(1) = O(n)
Space: O(n)

````python
nums=[2,7,11,15]
target=9
for i in range(len(nums)):
    for j in range(i+1,len(nums)):
        if nums[i]+nums[j]==target:
            print(i,j)
````

### Extension 5: Write an optimized solution

The HashMap solution improves time complexity from `O(n^2)` to `O(n)`.

````python
def two_sum_hashmap(nums, target):
    """
    Optimized HashMap solution for Two Sum.

    Time Complexity: O(n)
    Space Complexity: O(n)
    """

    # Basic quality control
    if not isinstance(nums, list):
        raise TypeError("nums must be a list")

    if len(nums) < 2:
        raise ValueError("nums must contain at least two elements")

    if not isinstance(target, Number):
        raise TypeError("target must be a number")

    for x in nums:
        if not isinstance(x, Number):
            raise TypeError("all elements in nums must be numbers")

    seen = {}

    for i, num in enumerate(nums):
        complement = target - num

        if complement in seen:
            return [seen[complement], i]

        seen[num] = i

    return None

two_sum_hashmap([2, 7, 11, 15], 9)
````
