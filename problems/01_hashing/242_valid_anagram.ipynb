# LeetCode 242. Valid Anagram

## Pattern

Hashing → Frequency Counting (HashMap / Dictionary)

---

## Problem

Given two strings `s` and `t`, return `True` if `t` is an anagram of `s`, and `False` otherwise.

An anagram contains exactly the same characters with the same frequencies.

---

## My First Thought

My first idea was to count how many times each character appears in both strings using a dictionary.

Then compare the two dictionaries.

If every character has the same frequency, the two strings are anagrams.

---

## Key Observation

Unlike **Contains Duplicate**, this problem is **not** asking whether a character has appeared before.

Instead, it asks:

> **How many times does each character appear?**

When a problem asks for frequencies or counts, a **dictionary (HashMap)** is usually the right data structure.

---

## Algorithm

1. If the two strings have different lengths, return `False`.
2. Create one dictionary for `s`.
3. Count the frequency of every character.
4. Repeat for `t`.
5. Compare the two dictionaries.

---

## Complexity

**Time Complexity**

`O(n)`

Both strings are traversed once.

**Space Complexity**

`O(n)`

In the worst case, every character is unique.

```python
a = "strings"

freq = {} #dict[ ]also works

for char in a:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1

print(freq)
```

```python
def isAnagram(s, t):

    if len(s) != len(t):
        return False

    freq_s = {}
    freq_t = {}

    for char in s:
        if char in freq_s:
            freq_s[char] += 1
        else:
            freq_s[char] = 1

    for char in t:
        if char in freq_t:
            freq_t[char] += 1
        else:
            freq_t[char] = 1

    return freq_s == freq_t
```

# Extension

This problem introduced another common use of a HashMap.

| Problem | Hash Structure | Stores |
|---------|----------------|--------|
| Two Sum | Dictionary | number → index |
| Contains Duplicate | Set | number |
| Valid Anagram | Dictionary | character → frequency |

The same data structure (HashMap) can solve different types of problems by storing different information.

General rule:

- Need fast lookup → Dictionary / Set
- Need to count frequency → Dictionary
- Need to know whether an element exists → Set
- Need to store extra information (index, count, etc.) → Dictionary

# Reflection

This problem reinforced the idea that a dictionary is not only used for lookup, but also for **frequency counting**.

Compared with the previous two problems:

- **Two Sum** taught me to store **value → index**.
- **Contains Duplicate** taught me to use a **set** for membership checking.
- **Valid Anagram** taught me to store **character → frequency**.
