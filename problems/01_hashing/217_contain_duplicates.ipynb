# LeetCode 217. Contains Duplicate

## Pattern

HashSet / Set

## Problem

Given an integer array `nums`, return `True` if any value appears at least twice in the array.

Return `False` if every element is distinct.


# Key Idea

For each number:

1. Check whether it is already in `seen`.
2. If yes, return `True`.
3. If no, add it to `seen`.
4. If the loop finishes, return `False`.

The key question is:

> Have I seen this number before?

When a problem asks this type of question, I should think about using a `set`.

````python
def containsDuplicate(nums):
    seen=set() 

    for num in nums:
        if num in seen:
            return True 
        seen.add(num)
    return False
nums=[1,2,3,1] 
print(containsDuplicate(nums))
````

# Complexity

## Time Complexity

`O(n)`

Reason:

We loop through the array once.

Checking whether `num in seen` is average `O(1)` because `seen` is a set.

So the total time complexity is: O(n)

Space Complexity

O(n)

Reason:

In the worst case, if all numbers are unique, we need to store all numbers in seen.

## 1. Use `for num in nums` when index is not needed

I first wrote:


- Rule:
If I need the index, use enumerate(nums).
If I only need the value, use for num in nums.

Use `set` when the question asks:

- Have I seen this value before?
- Are there duplicates?
- Does this value exist?
- Can I check membership quickly?

# Extension Questions

## Extension 1

If the question asks:

> Return the duplicate number.

Then I can still use a set.

```python
def findDuplicateValue(nums):
    seen = set()

    for num in nums:
        if num in seen:
            return num

        seen.add(num)

    return None
```


If the question asks:

Return the first index where a duplicate is found.

````python
def firstDuplicateIndex(nums):
    seen = set()

    for i, num in enumerate(nums):
        if num in seen:
            return i

        seen.add(num)

    return -1
````

If the question asks:

Return the original index of the duplicated value.

````python
def firstOriginalDuplicateIndex(nums):
    seen = {}

    for i, num in enumerate(nums):
        if num in seen:
            return seen[num]

        seen[num] = i

    return -1
````
