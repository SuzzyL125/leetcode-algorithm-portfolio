# LeetCode 49. Group Anagrams

## Pattern

HashMap → Grouping by Canonical Representation

---

## Problem

Given an array of strings `strs`, group the anagrams together.

Anagrams contain the same characters with the same frequencies, but possibly in different orders.

Example

Input

```python
["eat","tea","tan","ate","nat","bat"]
```

Output

```python
[
    ["eat","tea","ate"],
    ["tan","nat"],
    ["bat"]
]
```

# Algorithm

For every word:

1. Sort the characters.
2. Convert the sorted characters into a string.
3. Use the sorted string as the key.
4. If the key does not exist, create an empty list.
5. Append the original word into that list.
6. Return all groups.

````python
def groupAnagrams(strs):
    groups = {}
    for char in strs:
        a="".join(sorted(char))
        if a not in groups:
            groups[a]=[]
        groups[a].append(char)
    return list(groups.values())
print(groupAnagrams(strs))
````

defaultdict(list) 的意思是：

如果 key 不存在，自动帮我创建一个空 list。

````python
from collections import defaultdict

def groupAnagrams(strs):
    groups = defaultdict(list)

    for word in strs:
        key = "".join(sorted(word))
        groups[key].append(word)

    return list(groups.values())
print(groupAnagrams(strs))
````

# Complexity

Assume

- n = number of words
- k = average length of each word

## Time Complexity

Sorting one word takes

```
O(k log k)
```

Repeat for n words

```
O(n × k log k)
```

## Space Complexity

The dictionary stores every word once.

Additional space is approximately

```
O(n × k)
```

# Reflection

This problem introduced a new way to use a HashMap.

Previous problems

| Problem | Hash Structure | Stores |
|---------|----------------|--------|
| Two Sum | Dictionary | number → index |
| Contains Duplicate | Set | number |
| Valid Anagram | Dictionary | character → frequency |

This problem

```
sorted word → list of words
```

The biggest lesson is:

Instead of comparing objects one by one, try to convert them into a **canonical representation**.

If two objects have the same canonical representation, they belong to the same group.

Examples of canonical representations:

- Sorted string
- Frequency tuple
- Normalized form

This idea appears in many hashing and grouping problems.
